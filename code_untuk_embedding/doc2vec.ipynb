```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
import pandas as pd
import numpy as np

df = pd.read_csv('C:/Users/Geraldus Wilsen/Documents/ProjectRekomendasiObat/data/data_obat_fix_ordered.csv')
data = df['summary'].to_list()
```

```python
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(data)]
print(tagged_data)

max_epochs = 100
vec_size = 768
alpha = 0.025

model = Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm =1)
  
model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    #print('iteration {0}'.format(epoch))
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha

model.save("corpus_dense_doc2vec_768.model")
print("Model Saved")
```

```python
from gensim.models.doc2vec import Doc2Vec

# model= Doc2Vec.load("corpus_dense_doc2vec.model")
model= Doc2Vec.load("corpus_dense_doc2vec_768.model")
test_data = word_tokenize("obat tetes untuk menurunkan demam bayi dan anak".lower())
similar_doc = model.dv.most_similar(positive=[model.infer_vector(test_data)],topn=10)
result = []
for i in range (0,len(similar_doc)):
    result.append(int(similar_doc[i][0]))
    print(data[int(similar_doc[i][0])])

print(result)
```
